```python
# cijena = B0 + B1*visina + B2*golovi + B3*nastupi
# Y=cijena; x1=visina; x2=golovi; x3=nastupi
# B=(X'X)**(-1)*X'Y
import numpy as np
```

```python
Y = np.array([100, 55, 70, 35, 70, 95, 40, 50, 120, 80, 65, 110, 110, 55, 50])
```

```python
X = np.array([[1, 189, 37, 40], 
              [1, 175, 16, 26], 
              [1, 184, 22, 30], 
              [1, 165, 8, 16], 
              [1, 182, 24, 32], 
              [1, 188, 29, 38], 
              [1, 169, 13, 17], 
              [1, 172, 13, 22], 
              [1, 189, 39, 48], 
              [1, 184, 27, 35], 
              [1, 180, 20, 33], 
              [1, 191, 38, 47], 
              [1, 191, 34, 49], 
              [1, 172, 17, 27], 
              [1, 170, 15, 20]])
Xt = X.transpose()
```

```python
Bx= Xt@X

Bx_inv=np.linalg.inv(Bx)
By=Xt@Y
```

```python
B = Bx_inv@By
```

```python
import matplotlib.pyplot as plt
```

```python
visina = []
golovi = []
nastupi = []
for i in range (0,15):
    visina.append(X[i][1])
    golovi.append(X[i][2])
    nastupi.append(X[i][3])
```

```python
plt.scatter(golovi, Y)
plt.xlabel("golovi")
plt.ylabel("cijena")
plt.show()
```

```python
plt.scatter(nastupi, Y)
plt.xlabel("nastupi")
plt.ylabel("cijena")
plt.show()
```

```python
suma_Y = 0
suma_visina = 0
suma_golovi = 0
suma_nastupi = 0
for i in range(0,15):
    suma_Y += Y[i]
    suma_visina += visina[i]
    suma_golovi += golovi[i]
    suma_nastupi += nastupi[i]
ars_Y = suma_Y/15
ars_visina = suma_visina/15
ars_golovi = suma_golovi/15
ars_nastupi = suma_nastupi/15
```

```python
ro_brojnik_visina = 0
ro_brojnik_golovi = 0
ro_brojnik_nastupi = 0

ro_nazivnik_visina = 0
ro_nazivnik_golovi = 0
ro_nazivnik_nastupi = 0
ro_nazivnik_Y = 0

for i in range (0,15):
    ro_brojnik_visina += (visina[i]-ars_visina)*(Y[i]-ars_Y)
    ro_brojnik_golovi += (golovi[i]-ars_golovi)*(Y[i]-ars_Y)
    ro_brojnik_nastupi += (nastupi[i]-ars_nastupi)*(Y[i]-ars_Y)
    
    ro_nazivnik_visina += (visina[i]-ars_visina)**2
    ro_nazivnik_golovi += (golovi[i]-ars_golovi)**2
    ro_nazivnik_nastupi += (nastupi[i]-ars_nastupi)**2
    ro_nazivnik_Y += (Y[i]-ars_Y)**2
```

```python
import math
```

```python
ro_visina = ro_brojnik_visina / math.sqrt(ro_nazivnik_visina*ro_nazivnik_Y)
print(ro_visina)
```

```python
ro_golovi = ro_brojnik_golovi / math.sqrt(ro_nazivnik_golovi*ro_nazivnik_Y)
print(ro_golovi)
```

```python
ro_nastupi = ro_brojnik_nastupi / math.sqrt(ro_nazivnik_nastupi*ro_nazivnik_Y)
print(ro_nastupi)
```
